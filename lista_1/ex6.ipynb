Abaixo, segue a resolução do problema de agrupamento usando o algoritmo K-Means.

O agrupamento não ficou exatamente igual ao feito no Excel. Porém, é possível notar que alguns grupos que são mais evidentes visualmente ficaram iguais aos grupos da solução feita no Excel via PLIM.



```python
from sklearn.cluster import KMeans
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('df_kmeans.csv', sep=';')
x, y = df['x'], df['y']

kmeans = KMeans(n_clusters=15)
kmeans.fit(df)

plt.scatter(x, y, c=kmeans.labels_)
plt.show()
```
